Load sentiment & topic results

```python
import pandas as pd
from config import SENTIMENT_DATA, TOPIC_DATA

df_sent = pd.read_csv(SENTIMENT_DATA)
df_topic = pd.read_csv(TOPIC_DATA)

df_sent.head(), df_topic.head()
```

Merge datasets

```python
df_topic["row_id"] = df_topic.index
df_sent["row_id"] = df_sent.index

df = df_sent.merge(df_topic[["row_id", "topic"]], on="row_id", how="left")
df.head()
```

Topic distribution

```python
df["topic"].value_counts().plot(kind="bar", figsize=(6,4))
```

Sentiment distribution

```python
df["sentiment"].value_counts().plot(kind="bar", figsize=(6,4))
```

Save final merged analysis

```python
df.to_csv("final_analysis.csv", index=False)
df.head()
```
